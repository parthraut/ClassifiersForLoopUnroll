```python
import json
```

```python
with open("datasets/mehar_likes_baklava.json") as f:
    dataset = json.load(f)

first_program = next(iter(dataset))
first_value = dataset[first_program]
first_loop_info = first_value[0][0]
first_timing_info = first_value[1]["0"]

print(first_loop_info)
print(first_timing_info)

feature_names = list(first_loop_info.keys())
print(feature_names)
```

```python
import pandas as pd

df_columns = ["program_name", "optimal_luf"] + feature_names
df = pd.DataFrame(columns=df_columns)

for program_name in dataset:
    program_info = dataset[program_name]
    feature_info = program_info[0]
    timing_info = program_info[1]

    feature_info = sorted(feature_info, key=lambda x: x["loop_line_number"])
    timing_info = [timing_info[loop_idx] for loop_idx in timing_info]

    loop_line_numbers = [loop_feature_info["loop_line_number"] for loop_feature_info in feature_info]
    good_feature_pass = (len(loop_line_numbers) == len(set(loop_line_numbers)) and not 0 in loop_line_numbers and len(feature_info) == len(timing_info))
    if not good_feature_pass:
        print("=================")
        print("Bad loop line numbers from feature pass. Continuing...")
        continue

    for loop_idx in range(len(feature_info)):
        loop_feature_info = feature_info[loop_idx]
        loop_timing_info = timing_info[loop_idx]
        optimal_luf = int(min(loop_timing_info, key=loop_timing_info.get))

        loop_feature_info.update(
            {
                "program_name": program_name, 
                "optimal_luf": optimal_luf
            }
        )

        df.loc[len(df.index)] = loop_feature_info
```

```python
df = df.replace({False: 0, True: 1})
df.to_csv('dataset.csv', index=False)
df
```
